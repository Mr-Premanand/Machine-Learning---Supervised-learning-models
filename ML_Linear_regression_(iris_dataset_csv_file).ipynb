# MODEL -**1**

```python

## Step-2
# now we will define the dependent and independent variable i.e y=mx+c to find relation between sepal.lenght and sepal.width
#Dependent variable 
y=iris[["sepal.length"]]

#independent variable                               # here give [[]] double squ bracket to define dataframe otherwise it will take as array
x=iris[["sepal.width"]]
```

```python
## Step-3
# Now we have to split the data into train and test set 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)

# here we have given test size 0.3 i.e 30% data from dataset will be consider as test data
# as we used "train_test_split" it split the data into 4 catogory i.e independent variable train and test set i.e x  & dependent variable train and test set i.e y  
```

```python
# now check the sets 
# independent train set
x_train.head()
```

```python
# independent test set
x_test.head()
```

```python
# dependent train set
y_train.head()
```

```python
# dependent test set
y_test.head()
```

```python
##Step-4
# Now import linear regression model and fit trail data in that to get relation 
from sklearn.linear_model import LinearRegression  # here we can use " as lr" to define lr
lr=LinearRegression()
```

```python
# Fit the train data
lr.fit(x_train,y_train)
```

```python
##Step-5
# now we will test data over by using independet variable x_test to check result as per the y_test
# use above trained linear regression relation
y_pred=lr.predict(x_test)
y_pred
```

```python
#Now vie actual y_test values to check wheather y_pred values are come as per y_test
y_test.head()
```

```python
##Step-6
# as we seen here there are diff between the y_pred and y_test values and we called it residual i.e Residual=y_test-y_pred
# sumation of square of residual is called mean square error 
from sklearn.metrics import mean_squared_error
```

```python
# mean squre error of model 1
mean_squared_error(y_test,y_pred)
```

```python
# Here we can also check the acurracy of model
lr.score(x_test,y_test)
```

# MODEL-2 
 In this we will consider multiple independent variables 

i.e sepal.width, petal.length, petal.width to predict sepal.length

```python
## Step-1
# now we will define the dependent and independent variable i.e y=mx+c to find relation between sepal.lenght 
# and multiple independent variable i.e sepal.width, petal.length, petal.width to predict sepal.length
#Dependent variable 
y=iris[["sepal.length"]]

#Independent variables
x=iris[["sepal.width","petal.length", "petal.width"]]
```

```python
## Step-2
# Now we have to split the data into train and test set 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
## Step-3
# we have imported linearRegression Lib already above so we define it as notaion lr2 for Model-2
lr2=LinearRegression()
```

```python
# Fit the train data
lr2.fit(x_train,y_train)
```

```python
##Step-4
# now we will test data over by using independet variable x_test to check result as per the y_test
# use above trained linear regression relation
y_pred=lr2.predict(x_test)
y_pred
```

```python
#Now vie actual y_test values to check wheather y_pred values are come as per y_test
y_test.head()
```

```python
## Step- 5
# we already imported MSE lib so we will directly find the MSE
# mean squre error of model 2
mean_squared_error(y_test,y_pred)
```
